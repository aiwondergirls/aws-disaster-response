```python
import tqdm
```

```python
#getting geo fromat from usgs data into points 
points=df_geo_usgs['geometry']
```

```python
points_list=[] #this list will contain geo points
location_list=[] #will contain country names related to the point
for p in tqdm.tqdm(points.values):
    containing_countries = countries_df[countries_df['geometry'].contains(p)][['SOVEREIGNT','ADMIN']].values
    
    if len(containing_countries) > 0:
        points_list.append(p)
        location_list.append(containing_countries)
```

```python
len(location_list)
```

```python
arr=location_list

v=[]#souvereign
j=[]#admin
for i in arr:
    
    v.append(i.item(0))
    j.append(i.item(1))
```

```python
d={'points':points_list,'souvereign':v, 'admin':j}
usgs_countries=gpd.GeoDataFrame(d)
```

```python
usgs_countries.head()
```

```python
usgs_countries=usgs_countries.rename(columns={'points':'geometry'})
```

```python
df_countries_usgs = gpd.sjoin(usgs_countries, df_geo_usgs, how="inner", op='intersects')
```
